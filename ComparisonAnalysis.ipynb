# Comparison Analysis <a class="anchor" id="section3"></a>

The data comprises 303 observations of separate individuals from the Cleveland Clinic Foundation, Hungarian Clinic Foundation, and Switzerland Clinic Foundation. The data analyzes 13 features and 1 target attribute for the three seperate data sets. The outcome of the invasive coronary angiography reflects the presence or absence of coronary artery disease in the patient, with 0 signifying the absence of CHD and labels 1-4 representing the presence of CHD. The majority of research on this set of data has focused on attempting to discern presence (value 1,2,3,4) from absence (value 0). Six cardiologists at the institution of the Cleveland Clinic Foundation did these tests. Patients that felt comfortable enough to take part in this procedure did so, hence they can be seen as volunteers. “Initially, the dataset contains 76 features or attributes from 303 patients; however, published studies chose only 14 features that are relevant in predicting heart disease. Hence, here we will be using the dataset consisting of 303 patients with 14 features set.” (https://towardsdatascience.com/exploratory-data-analysis-on-heart-disease-uci-data-set-ae129e47b323). Therefore this is not the raw data collected but rather the data that is deemed to be useful for the predictions of CHD without conducting any invasive procedures. 

### Variable Breakdown <a class="anchor" id="section3_2"></a>


Discrete Variables: age, trestbps, chol, thalach                                                                                               
Continous Variables: oldpeak                                                                                                                   
Categorical Variables in data: sex, cp, fbs, restecg, exang, slope, ca, thal, num

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set(style="darkgrid")

data1 = pd.read_csv(r"C:\Users\alish\Documents\Python\USC\Data Sets\processed.cleveland.data.csv", na_values="?")
df1 = pd.DataFrame(data1)


data2 = pd.read_csv(r"C:\Users\alish\Documents\Python\USC\Data Sets\processed.hungarian.data.csv", na_values="?")
df2 = pd.DataFrame(data2)

data3 = pd.read_csv(r"C:\Users\alish\Documents\Python\USC\Data Sets\processed.switzerland.csv", na_values="?")
df3 = pd.DataFrame(data3)
```

### Sex Comparison <a class="anchor" id="section3_2"></a>

The sex of the individual is taken into account with 1 meaning male and 0 meaning female.

```python
df1['Key'] = 'Cleveland'
df2['Key'] = 'Hungarian'
df3['Key'] = 'Switzerland'

DF = pd. concat([df1, df2, df3]) [["sex", "Key"]]
DF.sex.replace([1, 0], ['male','female'], inplace=True)

#print (DF)
DFGroup = DF.groupby(by=["Key", "sex"]).size()

#print (DFGroup)
DFGPlot = DFGroup. unstack("Key").plot(kind="bar")
```

### num Comparison <a class="anchor" id="section3_2"></a>

With values 0,1,2,3,4 meaning "<50% diameter narrowing", ">50% diameter narrowing", ">50% diameter narrowing", and ">50% diameter narrowing" respectively.
```python
df1['Key'] = 'Cleveland'
df2['Key'] = 'Hungarian'
df3['Key'] = 'Switzerland'

DF = pd. concat([df1, df2, df3]) [["num", "Key"]]
DF.num.replace([0, 1, 2, 3, 4], ['<50% diameter narrowing','>50% diameter narrowing','>50% diameter narrowing', '>50% diameter narrowing', '>50% diameter narrowing'], inplace=True)

#print (DF)
DFGroup = DF.groupby(by=["Key", "num"]).size()

#print (DFGroup)
DFGPlot = DFGroup. unstack("Key").plot(kind="bar")
```
