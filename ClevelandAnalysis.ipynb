### Variable Breakdown <a class="anchor" id="section3_2"></a>

Discrete Variables: age, trestbps, chol, thalach                                                                                               
Continous Variables: oldpeak                                                                                                                   
Categorical Variables in data: sex, cp, fbs, restecg, exang, slope, ca, thal, num

```python
import pandas as pd
data = pd.read_csv(r"C:\Users\alish\Documents\Python\USC\Data Sets\processed.cleveland.data.csv", na_values="?")
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set(style="darkgrid")
df = pd.DataFrame(data)
```

### Cleveland Age Summary <a class="anchor" id="section3_2"></a>

The age of the indvidual will be taken into account for the data set.

```python
print('Cleveland age mean: ' + str(data.age.mean()))
print('Cleveland age median: ' + str(data.age.median()))
print('Cleveland age mode: ' + str(data.age.mode()))
print('Cleveland age range: ' + str(data.age.max() - data.age.min()))
print('Cleveland age min: ' + str(data.age.min()))
print('Cleveland age max: ' + str(data.age.max()))
print('Cleveland age #n: ' + str(data.age.count()))
print('Cleveland age Quartile 1: ' + str(np.percentile(data.age, 25)))
print('Cleveland age Quartile 2: ' + str(np.percentile(data.age, 50)))
print('Cleveland age Quartile 3: ' + str(np.percentile(data.age, 75)))
print('Cleveland age IQR: ' + str(np.percentile(data.age, 75)-np.percentile(data.age, 25)))
```

```python
import matplotlib.pyplot as plt

plt.hist(data['age'], ec="black")
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.show()
```
