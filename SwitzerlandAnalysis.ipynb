```python
import pandas as pd
data = pd.read_csv(r"C:\Users\alish\Documents\Python\USC\Data Sets\processed.switzerland.csv", na_values="?")
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set(style="darkgrid")
df = pd.DataFrame(data)
```

### Switzerland Age Number Summary <a class="anchor" id="section3_2"></a>

The age of the indvidual will be taken into account for the data set.

```python
print('Switzerland age mean: ' + str(data.age.mean()))
print('Switzerland age median: ' + str(data.age.median()))
print('Switzerland age mode: ' + str(data.age.mode()))
print('Switzerland age range: ' + str(data.age.max() - data.age.min()))
print('Switzerland age min: ' + str(data.age.min()))
print('Switzerland age max: ' + str(data.age.max()))
print('Switzerland age #n: ' + str(data.age.count()))
print('Switzerland age Quartile 1: ' + str(np.percentile(data.age, 25)))
print('Switzerland age Quartile 2: ' + str(np.percentile(data.age, 50)))
print('Switzerland age Quartile 3: ' + str(np.percentile(data.age, 75)))
print('Switzerland age IQR: ' + str(np.percentile(data.age, 75)-np.percentile(data.age, 25)))
```

```python
import matplotlib.pyplot as plt
plt.hist(data['age'], ec="black")
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.show()
```
