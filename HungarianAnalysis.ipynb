```python
import pandas as pd
data = pd.read_csv(r"C:\Users\alish\Documents\Python\USC\Data Sets\processed.hungarian.data.csv", na_values="?")
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set(style="darkgrid")
df = pd.DataFrame(data)
```

### Hungarian Age Number Summary <a class="anchor" id="section3_2"></a>

```python
print('Hungarian age mean: ' + str(data.age.mean()))
print('Hungarian age median: ' + str(data.age.median()))
print('Hungarian age mode: ' + str(data.age.mode()))
print('Hungarian age range: ' + str(data.age.max() - data.age.min()))
print('Hungarian age min: ' + str(data.age.min()))
print('Hungarian age max: ' + str(data.age.max()))
print('Hungarian age #n: ' + str(data.age.count()))
print('Hungarian age Quartile 1: ' + str(np.percentile(data.age, 25)))
print('Hungarian age Quartile 2: ' + str(np.percentile(data.age, 50)))
print('Hungarian age Quartile 3: ' + str(np.percentile(data.age, 75)))
print('Hungarian age IQR: ' + str(np.percentile(data.age, 75)-np.percentile(data.age, 25)))
```
